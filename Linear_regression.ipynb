#### Ex2: Viết code numpy, tìm model linear regression cho bai toán dữ đoán giá nhà, dataset:
#### a. Vẽ model dự đoán (đường thẳng) và dữ liệu (point - scatter).
#### b. Dự đoán giá các căn nhà có diện tích 50, 100, 150

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/nttuan8/DL_Tutorial/master/L1/data_linear.csv')
df
```

```python
df.shape
```

```python
x = df['Diện tích']
y = df['Giá']
plt.xlabel('Diện tích')
plt.ylabel('Giá')
plt.scatter(x,y)
```

```python
model = np.polyfit(x, y, 1)
model
```

```python
predict = np.poly1d(model)
```

```python
X = range(20, 100)
Y = predict(X)
plt.scatter(x, y)
plt.plot(X, Y, c = 'r')
```

```python
S = 50
print("Nhà có diện tích",S, "có giá là:", predict(S))
```

```python
S = 100
print("Nhà có diện tích",S, "có giá là:", predict(S))
```

```python
S = 150
print("Nhà có diện tích",S, "có giá là:", predict(S))
```
